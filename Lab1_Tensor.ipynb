```python
# Tạo tensor x kích thước (3, 2) với giá trị trong khoảng [5, 10]
x = torch.randint(5, 11, (3, 2))

# Tạo tensor y kích thước (2, 3) với giá trị trong khoảng [15, 20]
y = torch.randint(15, 21, (2, 3))

print("x:\n", x)
print("y:\n", y)

# Giữ giá trị đường chéo của x
diag_x = torch.diag(torch.diag(x))

print("Đường chéo của x:\n", diag_x)

# Để thực hiện x + y, cần điều chỉnh kích thước cho phù hợp
# Ví dụ: chuyển y thành (3, 2) bằng cách cắt hoặc reshape
y_resized = y.T  # chuyển vị để có (3, 2)

result = diag_x + y_resized

print("Kết quả x + y:\n", result)
```

```python
# Đặt tùy chọn hiển thị
torch.set_printoptions(precision=0)
# Hàm tạo 2 tensor ngẫu nhiên
def create_tensors():
    x = torch.randint(0, 10, (3, 2, 2), dtype=torch.float32)
    y = torch.randint(0, 10, (3, 2, 2), dtype=torch.float32)
    return x, y

# Các hàm toán học
def add_tensors(x, y):
    return x + y

def sub_tensors(x, y):
    return x - y

def mul_tensors(x, y):
    return x * y

def div_tensors(x, y):
    raw_result = x / y
    # Làm sạch kết quả: thay nan, inf bằng 0
    return torch.nan_to_num(raw_result, nan=0.0, posinf=0.0, neginf=0.0)

# Đưa các hàm này vào graph bằng TorchScript
scripted_add = torch.jit.script(add_tensors)
scripted_sub = torch.jit.script(sub_tensors)
scripted_mul = torch.jit.script(mul_tensors)
scripted_div = torch.jit.script(div_tensors)

# Thử chạy
x, y = create_tensors()
print("x:\n", x)
print("y:\n", y)

print("x + y:\n", scripted_add(x, y))
print("x - y:\n", scripted_sub(x, y))
print("x * y:\n", scripted_mul(x, y))
print("x / y (sạch):\n", scripted_div(x, y))
```

```python
# Định nghĩa model cộng hai tensor
class AddModel(torch.nn.Module):
    def __init__(self):
        super(AddModel, self).__init__()

    def forward(self, x, y):
        return x + y

# Khởi tạo model
model = AddModel()

# Tạo dữ liệu ngẫu nhiên
x = torch.randint(0, 10, (3, 2, 2), dtype=torch.float32)
y = torch.randint(0, 10, (3, 2, 2), dtype=torch.float32)

# Chạy model
output = model(x, y)

print("x:\n", x)
print("y:\n", y)
print("x + y:\n", output)
```

```python
# Định nghĩa model hiệu
class SubModel(torch.nn.Module):
    def __init__(self):
        super(SubModel, self).__init__()

    def forward(self, x, y):
        return x - y

# Định nghĩa model tích
class MulModel(torch.nn.Module):
    def __init__(self):
        super(MulModel, self).__init__()

    def forward(self, x, y):
        return x * y

# Khởi tạo model
sub_model = SubModel()
mul_model = MulModel()

# Tạo dữ liệu ngẫu nhiên
x = torch.randint(0, 10, (3, 2, 2), dtype=torch.float32)
y = torch.randint(0, 10, (3, 2, 2), dtype=torch.float32)

print("x:\n", x)
print("y:\n", y)

# Chạy model hiệu
print("x - y:\n", sub_model(x, y))

# Chạy model tích
print("x * y:\n", mul_model(x, y))
```

```python
class ParallelogramProofModel(torch.nn.Module):
    def __init__(self):
        super(ParallelogramProofModel, self).__init__()

    def forward(self, sides):
        # sides = [AB, BC, CD, DA]
        AB, BC, CD, DA = sides

        # Điều kiện: hai cặp cạnh đối bằng nhau
        cond_equal = (AB == CD) & (BC == DA)

        # Song song ⇔ bằng nhau (theo mệnh đề)
        cond_parallel = cond_equal

        return cond_parallel

# Ví dụ dữ liệu: hình bình hành với AB=4, BC=6, CD=4, DA=6
sides = torch.tensor([4.0, 6.0, 4.0, 6.0])
model = ParallelogramProofModel()
print("Mệnh đề đúng?", model(sides).item())
```

```python
class ParallelogramProofModel2(torch.nn.Module):
    def __init__(self):
        super(ParallelogramProofModel2, self).__init__()

    def forward(self, sides, parallel_AD_BC):
        AB, BC, CD, DA = sides

        # Mệnh đề: AB = CD ⇔ (AD = BC) + (AD // BC)
        lhs = (AB == CD)
        rhs = (DA == BC) & parallel_AD_BC

        return lhs == rhs

# Ví dụ dữ liệu: hình bình hành với AB=4, BC=6, CD=4, DA=6
sides = torch.tensor([4.0, 6.0, 4.0, 6.0])
parallel_AD_BC = True

model = ParallelogramProofModel2()
print("Mệnh đề đúng?", model(sides, parallel_AD_BC).item())
```

```python
class MidpointProofModel(torch.nn.Module):
    def __init__(self):
        super(MidpointProofModel, self).__init__()

    def forward(self, A, C, N):
        # A, C, N là tọa độ dạng tensor (2D hoặc 3D)
        # Kiểm tra N có phải trung điểm của AC
        midpoint = (A + C) / 2
        cond_midpoint = torch.allclose(N, midpoint)

        # Nếu N là trung điểm thì NA = NC
        NA = torch.norm(A - N)
        NC = torch.norm(C - N)
        cond_equal = torch.isclose(NA, NC)

        return cond_midpoint & cond_equal

# Ví dụ dữ liệu: A=(0,0), C=(2,0), N=(1,0)
A = torch.tensor([0.0, 0.0])
C = torch.tensor([2.0, 0.0])
N = torch.tensor([1.0, 0.0])

model = MidpointProofModel()
print("Mệnh đề NA = NC đúng?", model(A, C, N).item())
```
