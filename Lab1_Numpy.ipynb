```python
import numpy as np
import pandas as pd

# Tạo mảng 3x3 với số nguyên ngẫu nhiên từ 0 đến 20
arr = np.random.randint(0, 21, size=(3, 3))

print("Mảng:\n", arr)

# Kiểm tra tất cả phần tử khác 0 x``
result = np.all(arr != 0)

print("Tất cả phần tử khác 0:", result)
```

```python
a = np.array([1, 5, 10, 15])
b = np.array([2, 5, 8, 20])

print("a:", a)
print("b:", b)

print("greater (a > b):", np.greater(a, b))
print("greater_equal (a >= b):", np.greater_equal(a, b))
print("less (a < b):", np.less(a, b))
print("less_equal (a <= b):", np.less_equal(a, b))
```

```python
arr = np.random.randint(0, 21, 20)

print("Mảng ban đầu:", arr)

# Tạo điều kiện lọc
mask = (arr >= 9) & (arr <= 15)

# Đổi dấu các phần tử thỏa điều kiện
arr[mask] = -arr[mask]

print("Mảng sau khi đổi dấu:", arr)
```

```python
arr = np.random.randint(0, 10, (3, 3, 3))

print("Mảng 3x3x3:\n", arr)

# Tổng theo từng "ma trận con" (theo trục 0)
sum_axis0 = np.sum(arr, axis=0)

# Tổng theo từng dòng (theo trục 1)
sum_axis1 = np.sum(arr, axis=1)

# Tổng theo từng cột (theo trục 2)
sum_axis2 = np.sum(arr, axis=2)

print("\nTổng theo axis=0 (từng lớp):\n", sum_axis0)
print("\nTổng theo axis=1 (theo dòng):\n", sum_axis1)
print("\nTổng theo axis=2 (theo cột):\n", sum_axis2)
```

```python
# Tạo 2 Series
s1 = pd.Series([2, 4, 6, 8, 10])
s2 = pd.Series([1, 3, 5, 7, 10])

print("Series 1:\n", s1)
print("Series 2:\n", s2)

# Cộng
print("\nCộng:\n", s1 + s2)

# Trừ
print("\nTrừ:\n", s1 - s2)

# Nhân
print("\nNhân:\n", s1 * s2)

# Chia
print("\nChia:\n", s1 / s2)
```

```python
# Tạo 2 Series
sr1 = pd.Series([1, 2, 3, 4, 5])
sr2 = pd.Series([2, 4, 6, 8, 10])

print("sr1:\n", sr1)
print("\nsr2:\n", sr2)

# Lọc các phần tử của sr1 không có trong sr2
result = sr1[~sr1.isin(sr2)]

print("\nItems of sr1 not present in sr2:\n", result)
```

```python
# Tạo 2 Series ngẫu nhiên 20 phần tử (giá trị 0–30)
x = pd.Series(np.random.randint(0, 30, 20))
y = pd.Series(np.random.randint(0, 30, 20))

print("Series x:\n", x.values)
print("\nSeries y:\n", y.values)

# Phép hội (union)
union_result = np.union1d(x, y)

# Phép giao (intersection)
intersect_result = np.intersect1d(x, y)

print("\nUnion (Hội):\n", union_result)
print("\nIntersection (Giao):\n", intersect_result)
```
